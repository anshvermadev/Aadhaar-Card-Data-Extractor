```python
# train.py
# Run this on Kaggle with the aadhar-dataset attached
# Dataset: https://www.kaggle.com/datasets/nagendra048/aadhar-dataset

import os
from ultralytics import YOLO

# Path to data.yaml from the dataset
DATA_YAML = "/kaggle/input/datasets/nagendra048/aadhar-dataset/data.yaml"

# Load pretrained YOLOv8 small model
model = YOLO("yolov8s.pt")

# Train
results = model.train(
    data=DATA_YAML,
    epochs=100,
    imgsz=640,
    batch=16,
    name="aadhar_yolov8",
    patience=20,        # early stopping
    device=0,           # GPU (use 'cpu' if no GPU)
    workers=4,
    optimizer="AdamW",
    lr0=0.001,
    augment=True,
)

# Save trained model
model.save("yolov8_trained_model.pt")
print("✅ Model saved as yolov8_trained_model.pt")
print(f"Best mAP50: {results.results_dict.get('metrics/mAP50(B)', 'N/A')}")
```
